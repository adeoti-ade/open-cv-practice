```python
image_color = cv2.imread("CocaCola_900.jpeg", 1)
```

```python
plt.imshow(image_color)
```

```python
image_color_reversed = image_color[:, :, ::-1]
```

```python
plt.imshow(image_color_reversed)
```

```python
b,g,r = cv2.split(image_color)
```

```python
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(b, cmap="gray"),plt.title("Blue channel")
plt.subplot(142);plt.imshow(g, cmap="gray"),plt.title("Green channel")
plt.subplot(143);plt.imshow(r, cmap="gray"),plt.title("Red channel")
plt.subplot(144);plt.imshow(cv2.merge((b,g,r))[:,:,::-1]);plt.title("Merged")
```

```python
plt.imshow(image_color)
```

```python
image_bgr_to_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_bgr_to_rgb)
```

```python
image_hsv = cv2.cvtColor(image_color, cv2.COLOR_BGR2HSV)
```

```python
h, s, v = cv2.split(image_hsv)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h, cmap="gray"),plt.title("Hue channel")
plt.subplot(142);plt.imshow(s, cmap="gray"),plt.title("Saturation channel")
plt.subplot(143);plt.imshow(v, cmap="gray"),plt.title("value channel")
plt.subplot(144);plt.imshow(image_bgr_to_rgb);plt.title("Original")
```

```python
h_new = h - 10
image_hsv_modified = cv2.merge((h_new, s, v))
image_hsv_modified_2_rgb = cv2.cvtColor(image_hsv_modified, cv2.COLOR_HSV2RGB)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h, cmap="gray"),plt.title("Hue channel")
plt.subplot(142);plt.imshow(s, cmap="gray"),plt.title("Saturation channel")
plt.subplot(143);plt.imshow(v, cmap="gray"),plt.title("value channel")
plt.subplot(144);plt.imshow(image_hsv_modified_2_rgb);plt.title("Modified")
```

```python
plt.imsave("modified.jpeg", image_hsv_modified_2_rgb)
```

```python
cv2.imwrite("modified_less_hue.jpeg", image_hsv_modified_2_rgb)
```
